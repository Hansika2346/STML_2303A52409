Read the data with pandas and find features and target variables.
Normalize the data with min-max scaling
Split the data into train and test.:

```python
import pandas as pd
dp=pd.read_csv('/content/train.csv')
print(dp)
```

```python
y=dp['price_range']
print(y)
```

```python
x=dp.drop('price_range',axis=1)
print(x)
```

```python
d=x-x.min()/(x.max()-x.min())
d
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(d,y,test_size=0.2)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```
