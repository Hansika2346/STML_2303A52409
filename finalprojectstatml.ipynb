```python
import pandas as pd
import numpy as np
d=pd.read_csv('/content/tourism_dataset (2).csv')
print(d)
```

```python
d.isnull().sum()
```

```python
y=d['Accommodation_Available']
print(y)
```

```python
x=d.drop('Accommodation_Available',axis=1)
print(x)
```

```python
x.dtypes
```

```python
for column in x.columns:
    print(x[column].value_counts())
```

```python
for column in x.columns:
    print(x[column].unique())
```

```python

location_mapping = {location: i for i, location in enumerate(x['Location'].unique())}
country_mapping = {country: i for i, country in enumerate(x['Country'].unique())}
category_mapping = {category: i for i, category in enumerate(x['Category'].unique())}

x['Location'] = x['Location'].map(location_mapping)
x['Country'] = x['Country'].map(country_mapping)
x['Category'] = x['Category'].map(category_mapping)
```

```python
x.dtypes
```

```python
x.head()
```

```python
d.isnull().sum()
```

```python
y.replace(['Yes', 'No'], [1, 0], inplace=True)
print(y)
```

```python
Y=d['Accommodation_Available']
print(Y)
```

SPLITING

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python

print(Y_train)
```

```python
print(Y_test)
```

```python
y.value_counts()
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python

from sklearn.linear_model import LogisticRegression
lr2=LogisticRegression(max_iter=1000)
lr2.fit(X_train,Y_train)
```

```python
yp1=lr2.predict(X_test)
print(accuracy_score(Y_test,yp1))
```

```python
print(confusion_matrix(Y_test,yp1))
```

```python
print(classification_report(Y_test,yp1))
```

```python
##PERCEPTRON CLASSIFICATION
from sklearn.linear_model import Perceptron
lr1=Perceptron()
lr1.fit(X_train,Y_train)
```

```python
yp1=lr1.predict(X_test)
print(accuracy_score(Y_test,yp1))
```

```python
print(confusion_matrix(Y_test,yp1))
```

```python
print(classification_report(Y_test,yp1))
```

```python
##SVM CLASSIFICATION
from sklearn.svm import SVC
lr=SVC()
lr.fit(X_train,Y_train)
```

```python
yp=lr.predict(X_test)
print(accuracy_score(Y_test,yp))
```

```python
print(confusion_matrix(Y_test,yp))
```

```python
print(classification_report(Y_test,yp))
```
